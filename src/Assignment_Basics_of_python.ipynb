```python
def isPrimeNumber(number):
  if(number == 1):
    return False
  elif(number == 2):
    return True
  else:
    #Looping till half of the number to optimize the code
    for i in range(2, number//2):
      if(number % i == 0):
        return False
    return True

isPrimeNumber(9)
```

```python
import random
# Function to genrate two random number and check product
def random_number_product_ckeck():
  # Generate First Random number
  num1 = random.randint(0,20)
  # Generate Second Random number
  num2 = random.randint(0,20)
  # Product of First & Second Random number
  numbers_product = num1* num2

  try:
    # Product if number Input from user
    user_product_answer = int(input(f"Enter the product of {num1} and {num2} :"))
    if(user_product_answer == numbers_product):
      print("Correct Answer")
    else:
      print("Wrong Answer")
  except:
    print("Invalid input")


random_number_product_ckeck()
```

```python
# Function to genrate square of odd number between 100 to 200
def square_of_odd_numbers():
  for oddNum in range(101,200,2):
    print(f"The square of {oddNum} is {oddNum**2}")
square_of_odd_numbers()
```

```python
# Function to genrate square of Even number between 100 to 200
def square_of_even_numbers():
  for evenNum in range(100,200,2):
    print(f"The square of {evenNum} is {evenNum**2}")

square_of_even_numbers()
```

```python
# Function to count words
import re
def count_word(text_input):
  return len(re.findall("\S+",text_input))
count_word("This is Pthon basic assingment")
```

```python
# Function to check palindrome
import re
def is_palindrome(text_input):
  return text_input == text_input[::-1]

is_palindrome("racecar")
```
