```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import warnings
warnings.filterwarnings('ignore')
```

```python
### Assignment Tasks
### a. Build 99% Confidence Interval Using Sample Standard Deviation
```

```python
### To construct 99% CI fo the mean durability will use t-distribution beacuase
    ## 1. Standard Deviation is Unknown
    ## 2. Sample Size small (< 30)
    ## 3. The t-distribution accounts for extra variability in smaller samples.
```

```python
### Step-by-Step Solution
```

```python
## Given Data
durability_sample_data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
## Sample Size
n = len(durability_sample_data)
## Confidence level 99%
conf_level = 0.99
```

```python
### Step 1 - Compute Sample data Statistics

mean = np.mean(durability_sample_data)
std_dev = np.std(durability_sample_data, ddof=1)
```

```python
### Step 2 - Compute  t critical value
```

```python
##Degree of fredom
deg_freedom = n - 1

## t critical
t_critical = stats.t.ppf((1 + conf_level) / 2, deg_freedom)

```

```python
### Step 3 - Margin of error
margin = t_critical * (std_dev / np.sqrt(n))
```

```python
### Step 4: Construct Confidence Interval

# Confidence interval
ci_lower = mean - margin
ci_upper = mean + margin

print(f"99% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
### Final Answer:
    ## The 99% confidence interval for the mean durability of print-heads is approximately (1.09, 1.39) million characters.
```

```python
### Assignment Tasks
    ## b. Build 99% Confidence Interval Using Known Population Standard Deviation
    ## If it were known that the population standard deviation is 0.2 million characters, construct a 99% confidence interval for the mean number of characters printed before failure.
```

```python
### To construct 99% CI fo the mean durability will use t-distribution beacuase
    ## 1. Since the population standard deviation (Ïƒ) is now known, we can use the Z-distribution (normal distribution) instead of the t-distribution to construct the 99% confidence interval.
```

```python
std_dev = 0.2
```

```python
z_critical = stats.norm.ppf((1 + conf_level) / 2)
```

```python
z_critical_margin_error = z_critical * (std_dev / np.sqrt(n))
```

```python
# Confidence interval
ci_lower = mean - z_critical_margin_error
ci_upper = mean + z_critical_margin_error

print(f"99% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
### Final Answer:
### If the population standard deviation is known to be 0.2, then the 99% confidence interval for the mean durability of print-heads is approximately (1.115, 1.381) million characters
```
